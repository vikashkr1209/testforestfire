```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_CLEANED.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(["day","month","year"],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
#Encoding 
df['Classes']= np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df.head()
```

```python
df.tail()
```

```python
df['Classes'].value_counts()
```

```python
## Independent and Dependent Features
X = df.drop('FWI',axis=1)
y = df['FWI']
```

```python
X.head()
```

```python
y
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
# Feature selection 
X_train.corr()
```

```python
# Checking for multicollinearity
plt.figure(figsize=(10,6))
corr = X_train.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()





```

```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
## thesehold --Domain expertise
corr_features = correlation(X_train, 0.85)
```

```python
## drop the features when correlation is more than 0.85
X_train.drop(corr_features, axis=1, inplace=True)
X_test.drop(corr_features, axis=1, inplace=True)
```

```python
## feature scalling and standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scale = scaler.fit_transform(X_train)
X_test_scale = scaler.transform(X_test)
```

```python
X_train_scale
```

```python
## Box plot to undaerstand effect of standard scaler
```

```python
plt.subplots(figsize=(12,6))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title('Before Scalling')
plt.subplot(1,2,2)
sns.boxplot(data=X_train_scale)
plt.title('After Scalling')
# plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error
lr = LinearRegression()
lr.fit(X_train_scale,y_train)
y_pred = lr.predict(X_test_scale)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
print("Mean Absolute Error:",mae)
print("R2 Score:",score)
plt.scatter(y_test,y_pred)
```

```python
# lasso regression
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score, mean_absolute_error
lasso = Lasso()
lasso.fit(X_train_scale,y_train)
y_pred = lasso.predict(X_test_scale)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
print("Mean Absolute Error:",mae)
print("R2 Score:",score)
plt.scatter(y_test,y_pred)
```

```python
# ridge regression
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score, mean_absolute_error
ridge = Ridge()
ridge.fit(X_train_scale,y_train)
y_pred = ridge.predict(X_test_scale)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
print("Mean Absolute Error:",mae)
print("R2 Score:",score)
plt.scatter(y_test,y_pred)
```

```python
# elastic net regession
from sklearn.linear_model import ElasticNet
from sklearn.metrics import r2_score, mean_absolute_error
elastic = ElasticNet()
elastic.fit(X_train_scale,y_train)
y_pred = elastic.predict(X_test_scale)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
print("Mean Absolute Error:",mae)
print("R2 Score:",score)
plt.scatter(y_test,y_pred)
```
