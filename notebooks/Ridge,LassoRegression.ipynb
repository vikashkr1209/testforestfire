```python
## fix spaces in column names 
df.columns=df.columns.str.strip()
```

```python
## chnage the required column as integer data type


df[['month','day','year','Temperature','RH','Ws']] = df[['month','day','year','Temperature','RH','Ws']].astype('int')
```

```python
#### changing other columns to float data type
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype('float')
```

```python
# Clean the text (remove spaces, lowercase, normalize multiple spaces)
df_copy['Classes'] = df_copy['Classes'].astype(str).str.strip().str.lower()
df_copy['Classes'] = df_copy['Classes'].str.replace(r'\s+', ' ', regex=True)

# Encode (not fire → 0, fire → 1)
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire', na=False), 0, 1)
```
